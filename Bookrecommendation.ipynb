```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'numOfRating'},inplace=True)
```

```python
number_rating.head()
```

```python
final=ratings_with_books.merge(number_rating,on='title')
```

```python
final.head()
```

```python
final.shape
```

```python
final=final[final['numOfRating']>=50]
```

```python
final.head()
```

```python
final.shape
```

```python
final.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final.shape
```

```python
book_pivot = final.pivot_table(columns='user_id', index='title', values= 'rating')
```

```python
print(book_pivot.shape)
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
book_pivot.head() 	
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
print(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
#return the distances- return the distnce betwwwnn book in 237th index and other books
#suggestion: index of top 6 near books
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6 )
```

```python
distance
```

```python
suggestion
book_names = book_pivot.index
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
book_name = []
for book_id in suggestion:
    book_name.append(book_pivot.index[book_id])
```

```python
book_name
```

```python
ids_index = []
for name in book_name[0]:
    ids = np.where(final['title'] == name)[0][0]
    ids_index.append(ids)
```

```python
for idx in ids_index:
    url = final.iloc[idx]['image_url']
    print(url)
```

```python
import pickle
pickle.dump(model,open('artifacts/model.pkl','wb'))
pickle.dump(book_names,open('artifacts/book_names.pkl','wb'))
pickle.dump(final,open('artifacts/final_rating.pkl','wb'))
pickle.dump(book_pivot,open('artifacts/book_pivot.pkl','wb'))
```

```python
# testing phase
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=4)
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"You searched '{book_name}'\n")
                print("The suggestion books are: \n")
            else:
                print(j)
```
